```python
df_y = df[['rank', 'query_id']].values
#df_qid = df['query_id'].values
df_x = df.drop(['rank', 'query_id'], axis=1)
```

```python
df_x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train_, y_test_ =  train_test_split(df_x, df_y)
```

```python
qid_train = pd.DataFrame(y_train_)[1].values
y_train = pd.DataFrame(y_train_)[0].values
qid_test = pd.DataFrame(y_test_)[1].values
y_test = pd.DataFrame(y_test_)[0].values
X_train = X_train.values
X_test = X_test.values
```

```python
indx = np.argsort(qid_train)
indx_test = np.argsort(qid_test)
```

```python
qid_train = qid_train[indx]
y_train = y_train[indx]
X_train = X_train[indx]
```

```python
qid_test = qid_test[indx_test]
y_test = y_test[indx_test]
X_test = X_test[indx_test]
```

```python
import xgboost as xgb
```

```python
print(X_train.shape, y_train.shape, qid_train.shape)
```

```python
ranker = xgb.XGBRanker(tree_method="hist", lambdarank_num_pair_per_sample=8, objective="rank:ndcg", lambdarank_pair_method="topk")
ranker.fit(X_train, y_train, qid=qid_train)
```

```python
scores = ranker.predict(X_test)
sorted_idx = np.argsort(scores)[::-1]
# Sort the relevance scores from most relevant to least relevant
#scores = scores[sorted_idx]
```
